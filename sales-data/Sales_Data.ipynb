Exploring Data

```python
df = pd.read_csv("C:\\Users\\LENOVO\\pandas\\sales\\orders.csv",skiprows= 4)

# df.info()
df.head(20)
# df.tail()


```

Drop Null Values

df.dropna(how = "all")

```python
df.dropna(how= "all",inplace=True)
df
```

Modify Email Column to be lower case

```python
df["Email"] = df["Email"].str.lower()
df.dropna()
df = df
```

Modify PhoneNumber to Replace "Tel:" with ""

```python
df["Phone Number"] = df["Phone Number"].str.replace("Tel:","")
df
```

Add FullName as New Column

```python


df["Customer FullName"] = df["Customer First Name"].str.title() + " " + df["Customer Last Name"].str.title()
df.dropna(how = "any")
df.dropna(how = "all",inplace=True)
fullName = df.columns[-1]
column_to_move = df.pop(fullName)
df.insert(9,fullName,column_to_move)

df.head(20)
```

Clean phone number coulmn

```python
df["Phone Number"] = df["Phone Number"].str.replace("Tel","")
df
```

Add Price and Final Price Coulmns

```python
df["Price"] = df["Quantity"] * df["Unit Price"]
df["Discount"].fillna(0,inplace =True)
df["Final Price"] = df["Quantity"] * df["Unit Price"] - df["Discount"]
df
```

Add Order Status Column (with value of Delevered or Pending)

```python
df.loc[df["Status"]== True,"OrderStatus"] = "Delevered"
df.loc[df["Status"]== False,"OrderStatus"] = "Pending"

df
```
